```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(
    'API_SP.POP.TOTL_DS2_en_csv_v2_7.csv',
    skiprows=4
)

data.head()
```

```python
data.columns
```

```python
population_2022 = data[['Country Name', '2022']].dropna()
population_2022.head()
```

```python
top10 = population_2022.sort_values(by='2022', ascending=False).head(10)
top10
```

```python
plt.figure(figsize=(10,5))
plt.bar(top10['Country Name'], top10['2022'])

plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Most Populated Countries (2022)')
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(population_2022['2022'], bins=10)

plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.title('Population Distribution of Countries (2022)')

plt.show()
```
